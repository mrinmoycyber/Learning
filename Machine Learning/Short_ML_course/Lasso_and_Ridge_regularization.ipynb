```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
# let's use limited columns which makes more sense for serving our purpose
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount',
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
```

```python
dataset.head()
```

```python
dataset.shape
```

Checking for Nan values

```python
dataset.isna().sum()
```

```python
# Some feature's missing values can be treated as zero (another class for NA values or absence of that feature)
# like 0 for Propertycount, Bedroom2 will refer to other class of NA values
# like 0 for Car feature will mean that there's no car parking feature with house
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)

# other continuous features can be imputed with mean for faster results since our focus is on Reducing overfitting
# using Lasso and Ridge Regression
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
dataset.head()
```

Let's one hot encode the categorical features

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset = dataset.astype(int)
```

```python
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

Here training score is 68% but test score is 13.85% which is very low

Normal Regression is clearly overfitting the data.

Alpha (alpha):
alpha is the regularization parameter in Lasso regression.
It controls the strength of the regularization penalty applied to the coefficients.
Higher values of alpha lead to more regularization (shrinkage) of the coefficients, potentially resulting in a simpler model with fewer features having non-zero coefficients.
Lower values of alpha reduce the amount of regularization, allowing more flexibility in the model to fit the training data.
It's typically chosen through techniques like cross-validation to find the optimal balance between bias and variance.
In your example, alpha=50 indicates a relatively high regularization strength.

Max Iterations (max_iter):
max_iter specifies the maximum number of iterations for which the algorithm should run.
Lasso regression is solved using iterative numerical methods (such as coordinate descent), and max_iter controls how many iterations these methods can take before stopping.
If convergence (optimization of coefficients) is not achieved within max_iter iterations, the algorithm stops and may return an incomplete solution.
In your example, max_iter=100 means that the algorithm will attempt to converge on the optimal coefficients in at most 100 iterations.

Tolerance (tol):
tol (short for tolerance) is a stopping criterion.
It specifies the tolerance (precision) with which the solution is considered to be converged.
When the updates to the coefficients are smaller than tol during an iteration, the algorithm is assumed to have converged and stops early.
A smaller tol value requires stricter convergence criteria and may lead to more iterations, but it ensures a more precise solution.

```python
#Using Lasso (L1 Regularized) Regression model
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
#Using Ridge (L2 Regularized) Regression Model
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```

L1 regularization (Lasso) is good for picking important features and making simpler models with fewer features. L2 regularization (Ridge) helps make the model more stable and reduces issues with variables being highly correlated.
