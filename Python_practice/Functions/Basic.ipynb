```python
# positonal arguments
# positional arguments need to pass in same order
def volume(length, breadth, height):
    vol = length * breadth * height
    return vol
v = volume(5, 6, 7)
print(v)
```

```python
# keyword arguments
# keyword arguments can pass in any order
def volume(length, breadth, height):
    vol = length * breadth * height
    return vol
v = volume(length = 5, height = 6, breadth = 7)
print(v)
```

```python
# positional argumnet / keyword argumnet
def num(a, b, /, c, d):
    print(a, b, c, d)
    
num(5, 6, c=19, d=5)
```

```python
def num(a, b, c, d, /):
    print(a, b, c, d)
    
num(5, 6, 7, 5)
```
