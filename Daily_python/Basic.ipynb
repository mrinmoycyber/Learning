```python
#duplicate array 
def find_dup(arr):
    arr.sort()
    duplicates = []
    
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1] and arr[i] not in duplicates:
            duplicates.append(arr[i])
    return duplicates

arr = [1, 2, 3, 3, 5, 6, 7, 8, 8, 0.1, 0.1]
print(find_dup(arr))
```

```python
# filter even numbers
numbers = [1, 3, 4, 7, 6, 8, 9, 18, 23, 24.1, 26]
even = []
for x in numbers:
    if x % 2 == 0:
        even.append(x)
print(even)
```

```python
numbers = [1, 3, 4, 7, 6, 8, 9, 18, 23, 2]
odd = []
for x in numbers:
    if x % 2 != 0:
        odd.append(x)
print(odd)
    
```

```python
# A lambda function in Python is a small anonymous function written in a single line.
add = lambda a, b: a + b
print(add(2, 3))
```

```python
def add(a, b):
    return a + b
print(add(2, 3))
```
